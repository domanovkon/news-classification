```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from ipywidgets import Select, SelectMultiple, interactive,  Layout, Box, Label

df = pd.read_csv('testPredictions.csv', names=['description','true_category','predicted_category'])
df.head()

encoder = LabelEncoder()
df['categoryEncoded'] = encoder.fit_transform(df['true_category'])
df.describe()
```

```python
#выбрать столбец, который будет отвечать за тип
yselect = Select(
        options=list(df.columns),
        value=list(df.columns)[0],
        description='Select file:',
        disabled=False
    )
yselect = "true_category"
```

```python
#выбрать столбец, который будет отвечать за текст
xselect = Select(
        options=list(df.columns),
        value=list(df.columns)[0],
        description='Select file:',
        disabled=False
    )
xselect = "description"
```

```python
from matplotlib import pyplot as plt

#просто смотрим какие у нас есть лейблы
fig, ax = plt.subplots()
fig.suptitle("this is what we have", fontsize=12)
df[yselect].reset_index().groupby(yselect).count().sort_values(by=
       "index").plot(kind="barh", legend=False,
        ax=ax).grid(axis='x')
plt.show()
```
